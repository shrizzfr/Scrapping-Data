# WEBSCRAPPING from website  + Pandas


```python
from bs4 import BeautifulSoup
import requests
```

```python
url = 'https://en.wikipedia.org/wiki/List_of_English_football_champions'
page = requests.get(url)
soup = BeautifulSoup(page.text,'html')
```

```python
print(soup)
```

```python
table=soup.find('table',class_='wikitable sortable')

```

```python
table = soup.find_all('table')[3]
print(table)
```

```python
premier_league_titles=table.find_all('th')
```

```python
premier_league_titles
```

```python
pl_all_titles =[title.text.strip()for title in premier_league_titles]
print(pl_all_titles)
```

```python
import pandas as pd
```

```python
df = pd.DataFrame(columns = pl_all_titles)
df
```

```python
column_data = table.find_all('tr')
```

```python
for row in column_data[1:]:
    row_data = row.find_all('td')
    individual_row_data=[data.text.strip()for data in row_data]
    
    
    length =len(df)
    df.loc[length]= individual_row_data
```
